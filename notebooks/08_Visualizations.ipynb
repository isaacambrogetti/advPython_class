#### Bar Plot

```python
def get_students_df():
    data = {
        2017: {"BSc": 9262,  "MSc": 6158, "PhD": 4092},
        2018: {"BSc": 9517,  "MSc": 6590, "PhD": 4175},
        2019: {"BSc": 9895,  "MSc": 7037, "PhD": 4168},
        2020: {"BSc": 10355, "MSc": 7790, "PhD": 4316},
        2021: {"BSc": 10642, "MSc": 8299, "PhD": 4460},
        2022: {"BSc": 10665, "MSc": 8737, "PhD": 4561},
        2023: {"BSc": 10948, "MSc": 9018, "PhD": 4425},
    }

    return (
        pd.DataFrame.from_dict(data, orient="index")
        .rename_axis("Year")
        .reindex(columns=["BSc", "MSc", "PhD"])
    )
```

```python
df = get_students_df()

colors = ["#1b4f72", "#4e79a7", "#76b7b2"]

fig, ax = plt.subplots(figsize=(7, 6))

years = df.index
bsc = df["BSc"]
msc = df["MSc"]
phd = df["PhD"]

ax.bar(years, bsc, label="BSc", color=colors[0])
ax.bar(years, msc, bottom=bsc, label="MSc", color=colors[1])
ax.bar(years, phd, bottom=bsc + msc, label="PhD", color=colors[2])

ax.set(
    title="Students per Program Type (2017–2023)",
    xlabel="Year",
    ylabel="Number of Students",
)
ax.legend(
    title="Program Type",
    loc="upper center",
    bbox_to_anchor=(0.5, -0.25),
    ncol=3,
    frameon=True,
)
ax.grid(axis="y", linestyle="--", alpha=0.4)
ax.tick_params(axis="x", rotation=0)

fig.tight_layout()
plt.show()
```

### Line Plots

```python
# Specific example data: weeks (x) and number of passengers (y)
df = pd.DataFrame({
    "Week": [1, 2, 3, 4, 5, 6, 7, 8],
    "Airline A": [1200, 1350, 1280, 1500, 1600, 1550, 1700, 1800],
    "Airline B": [1000, 1100, 1080, 1250, 1300, 1400, 1450, 1500],
    "Airline C": [1400, 1450, 1500, 1650, 1750, 1720, 1850, 1950],
})

plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots(figsize=(9, 6))

for col in ["Airline A", "Airline B", "Airline C"]:
    ax.plot(
        df["Week"], df[col],
        label=col,
        linewidth=2.5,
        marker="o",
        markersize=5
    )

ax.set_title("Weekly Passenger Trend by Airline", fontweight="bold", pad=12)
ax.set_xlabel("Week")
ax.set_ylabel("Passengers (people)")

ax.set_xticks(df["Week"])  # show every week on the x-axis
ax.legend(title="Airline", frameon=True)

ax.grid(True, linestyle="--", alpha=0.35)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
```

### Scatterplot

```python
df = sns.load_dataset("penguins")

fig, ax = plt.subplots(figsize=(7, 5))

markers = {"Male": "o", "Female": "x"}

colors = {
    "Adelie": "tab:blue",
    "Chinstrap": "tab:orange",
    "Gentoo": "tab:green"
}

for species in df["species"].dropna().unique():
    for sex in df["sex"].dropna().unique():
        subset = df[(df["species"] == species) & (df["sex"] == sex)]

        ax.scatter(
            subset["flipper_length_mm"],
            subset["body_mass_g"],
            label=f"{species} - {sex}",
            marker=markers.get(sex, "o"),
            color=colors.get(species),
            alpha=0.8
        )

ax.set(
    title="Penguin Body Mass vs. Flipper Length",
    xlabel="Flipper Length (mm)",
    ylabel="Body Mass (g)",
)

ax.grid(True, linestyle="--", alpha=0.3)
ax.legend()

fig.tight_layout()
```

#### Seaborn Version

### Box Plots

```python
# Load data

df_ba = pd.read_csv(
    "../data/csv/basel_2024.csv",
    parse_dates=["date"],
    index_col="date",
).assign(station="Basel")

df_jf = pd.read_csv(
    "../data/csv/jungfraujoch_2024.csv",
    parse_dates=["date"],
    index_col="date",
).assign(station="Jungfraujoch")

combined = pd.concat(
    [
        df_ba.assign(station="Basel")[["tavg", "station"]],
        df_jf.assign(station="Jungfraujoch")[["tavg", "station"]],
    ],
    ignore_index=True,
)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

groups = [combined[combined["station"] == s]["tavg"].dropna()
          for s in ["Basel", "Jungfraujoch"]]

labels = ["Basel", "Jungfraujoch"]
colors = ["#a6cee3", "#fdbf6f"]

bp = ax.boxplot(
    groups,
    tick_labels=labels,
    widths=0.4,
    patch_artist=True,   
    showfliers=True,
    flierprops=dict(markersize=2),
    boxprops=dict(linewidth=1.2),
    capprops=dict(linewidth=1.2),
    whiskerprops=dict(linewidth=1.2),
    medianprops=dict(linewidth=1.2, color="black"),
)

# Farben setzen
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)

ax.set(
    title="Temperature Distribution per Station",
    xlabel="",
    ylabel="Daily Average Temperature (°C)",
)

ax.grid(True, linestyle="--", alpha=0.3)

fig.tight_layout()
```

#### Seaborn Version

```python
sns.set_theme(style="whitegrid", palette="pastel")

palette = {"Basel": "#a6cee3", "Jungfraujoch": "#fdbf6f"}

fig, ax = plt.subplots(figsize=(6, 4))
sns.boxplot(
    data=combined,
    x="station",
    y="tavg",
    hue="station",       # enables per-station colors
    palette=palette,
    width=0.4,
    showcaps=True,
    fliersize=2,
    linewidth=1.2,
    dodge=False,         # important when x == hue
    ax=ax,
)

# Remove redundant legend created by hue=station
leg = ax.get_legend()
if leg is not None:
    leg.remove()

ax.set(
    title="Temperature Distribution per Station",
    xlabel="",
    ylabel="Daily Average Temperature (°C)",
)
ax.grid(True, linestyle="--", alpha=0.3)

fig.tight_layout()
```

### Violin Plots

```python
sns.set_theme(style="whitegrid", palette="pastel")

palette = {"Basel": "#a6cee3", "Jungfraujoch": "#fdbf6f"}

fig, ax = plt.subplots(figsize=(6, 4))
sns.violinplot(
    data=combined,
    x="station",
    y="tavg",
    hue="station",      # required for explicit palette mapping in seaborn >= 0.14
    palette=palette,
    inner="box",
    linewidth=1.2,
    ax=ax,
)

# Remove redundant legend created by hue=station (since x == hue)
leg = ax.get_legend()
if leg is not None:
    leg.remove()

ax.set(
    title="Temperature Distribution per Station",
    xlabel="",
    ylabel="Daily Average Temperature (°C)",
)
ax.grid(True, linestyle="--", alpha=0.3)

fig.tight_layout()
```

<hr style="border: none; height: 5px; background-color: gray;">

### Violin-Swarm Plots

```python
sns.set_theme(style="whitegrid", palette="pastel")

palette = {"Basel": "#a6cee3", "Jungfraujoch": "#fdbf6f"}

fig, ax = plt.subplots(figsize=(6, 4))

# Background violin plot
sns.violinplot(
    data=combined,
    x="station",
    y="tavg",
    hue="station",  
    palette=palette,
    inner=None,
    linewidth=0,
    cut=0,
    ax=ax,
)

# Foreground swarm plot (raw observations)
sns.swarmplot(
    data=combined,
    x="station",
    y="tavg",
    size=2.5,
    color="black",
    alpha=0.6,
    ax=ax,
)

# Remove redundant legend (x == hue)
leg = ax.get_legend()
if leg is not None:
    leg.remove()

ax.set(
    title="Temperature Distribution per Station",
    xlabel="",
    ylabel="Daily Average Temperature (°C)",
)
ax.grid(True, linestyle="--", alpha=0.3)

fig.tight_layout()
```

### Histograms

```python
fig, ax = plt.subplots(figsize=(7, 5))

stations = combined["station"].dropna().unique()
colors = {"Basel": "#a6cee3", "Jungfraujoch": "#fdbf6f"}

for station in stations:
    subset = combined[combined["station"] == station]["tavg"].dropna()
    
    ax.hist(
        subset,
        bins=15,
        density=True, 
        alpha=0.5,
        label=station,
        color=colors.get(station),
        edgecolor=None
    )

ax.set(
    title="Distribution of Daily Average Temperature",
    xlabel="Daily Average Temperature (°C)",
    ylabel="Probability",
)

ax.grid(True, linestyle="--", alpha=0.3)
ax.legend()

fig.tight_layout()
```

#### Seaborn Version

```python
sns.set_theme(style="whitegrid", palette="pastel")

fig, ax = plt.subplots(figsize=(7, 5))
sns.histplot(
    data=combined,
    x="tavg",
    hue="station",
    bins=15,
    stat="probability",   # y-axis is probability (not counts)
    common_norm=False,    # normalize each station separately
    kde=False,            # set to True to overlay a kernel density estimate (smoothed curve)
    alpha=0.5,
    edgecolor=None,
    ax=ax,
)

ax.set(
    title="Distribution of Daily Average Temperature",
    xlabel="Daily Average Temperature (°C)",
    ylabel="Probability",
)
ax.grid(True, linestyle="--", alpha=0.3)

fig.tight_layout()
```

<hr style="border: none; height: 5px; background-color: gray;">

### Error Bars



```python
sns.set_theme(style="whitegrid", palette="pastel")

df = sns.load_dataset("penguins")

grouped = df.groupby("species", observed=True)["body_mass_g"].agg(mean="mean", std="std")

fig, ax = plt.subplots(figsize=(9, 6))
ax.errorbar(
    x=grouped.index,
    y=grouped["mean"],
    yerr=grouped["std"],
    fmt="o",
    color="#0a3e62",   # marker color
    ecolor="#1f78b4",  # error bar color
    elinewidth=2,
    capsize=5,
)

ax.set(
    title="Mean Body Mass per Penguin Species\nwith Standard Deviation as Error Bars",
    ylabel="Body Mass (g)",
)
fig.tight_layout()
```

### Uncertainty - Filled Area

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
err = 0.3

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label="Signal")
ax.fill_between(x, y - err, y + err, alpha=0.3, label="±0.3 range")

ax.set(
    title="Signal with Uncertainty Band (fill_between)",
    xlabel="x",
    ylabel="sin(x)",
)
ax.grid(True, linestyle="--", alpha=0.5)
ax.legend()

fig.tight_layout()
```

### Confidence Interval

```python
np.random.seed(42)

x = np.arange(0, 20)
y = 2.0 * x + 5 + np.random.normal(loc=0, scale=8, size=len(x))
df = pd.DataFrame({"x": x, "y": y})

sns.set_theme(style="whitegrid", context="talk")

fig, ax = plt.subplots(figsize=(10, 7))

sns.regplot(
    data=df, x="x", y="y",
    ci=95,
    line_kws=dict(linewidth=3),
    ax=ax
)

ax.set(
    title="Regression Trend Line with 95% Confidence Interval Band",
    xlabel="X",
    ylabel="Y",
)

plt.tight_layout();
```

# Visualizing Uncertainty

#### Error bars

```python
x = np.linspace(0, 10, 20)
y = np.sin(x) + np.random.randn(20)

# errors
yerr_lower = 0.3 + 0.2 * np.random.rand(20)
yerr_upper = 0.5 + 0.3 * np.random.rand(20)

yerr = [yerr_lower, yerr_upper]

fig, ax = plt.subplots()

ax.errorbar(
    x, y,
    yerr=yerr,
    fmt='ok',
    capsize=3
)
```

```python
fig, ax = plt.subplots()

ax.errorbar(
    x, y,
    yerr=yerr,
    marker='o',
    markersize=6,
    linestyle='',
    color='black',
    ecolor='lightgray',
    elinewidth=3,
    capsize=5,
    capthick=2,
    label='Error'
);
```

### Continuous Errors

```python
x = np.linspace(0, 2 * np.pi, 250)
random_delta = 0.5 + 0.2 * np.random.rand(len(x))

mean_values = np.sin(x) + 0.15 * np.random.randn(len(x))

upper_error_bound = mean_values + random_delta
lower_error_bound = mean_values - random_delta

fig, ax = plt.subplots()

ax.plot(x, mean_values, label="Mean signal")

ax.fill_between(
    x,
    lower_error_bound,
    upper_error_bound,
    alpha=0.2,
    label="Error band"
)

ax.set(
    title="Continuous Errors: Mean Signal with Uncertainty Band",
    xlabel="x",
    ylabel="Value"
)

ax.legend();
```

```python
import wfdb
```

```python
# Load the record (reads .hea and .dat)
record = wfdb.rdrecord('../data/ecg_db/patient001/s0010_re')

# Access all channels as a NumPy array
record.p_signal[0:20]

print(f"Channels ({len(record.sig_name)}): {record.sig_name}")
print(f"Sampling rate (Hz): {record.fs}")
```

### Preparations: Select 12 derivatives + define time window

```python
start_idx = 1300
end_idx = 2600

channels = record.p_signal.T  # Transpose: shape (channels, samples)
selected_leads = ['i', 'ii', 'iii', 'avr', 'avl', 'avf', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6']

lead_indices = [i for i, name in enumerate(record.sig_name) if name in selected_leads]
channels_12 = channels[lead_indices, :]  # Shape: (12, samples)
lead_names = [record.sig_name[i] for i in lead_indices]

print("Selected leads:", lead_names)
print("channels_12 shape:", channels_12.shape)
```

```python
fs = record.fs
colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728",
    "#9467bd", "#8c564b", "#e377c2", "#7f7f7f",
    "#bcbd22", "#17becf", "#393b79", "#637939"
]

fontsize = 20
x_offset = 40  # in samples
pad_s = x_offset / fs

segment = channels_12[:, start_idx:end_idx]
t = np.arange(start_idx, end_idx) / fs

fig, ax = plt.subplots(figsize=(15, 10))

for name, signal, color in zip(lead_names, segment, colors):
    ax.plot(t, signal, label=name, color=color)

ax.set(
    xlabel="Time (s)",
    ylabel="Amplitude (mV)",
    title="ECG Signal - 12 Leads – Patient 001",
    xlim=(t[0] - pad_s, t[-1] + pad_s),
)

ax.tick_params(labelsize=fontsize)
ax.minorticks_on()
ax.grid(True, which="major", linestyle="-", linewidth=0.8)
ax.grid(True, which="minor", linestyle=":", linewidth=0.6, alpha=0.7)

ax.legend(
    loc="upper right",
    ncol=2,
    fontsize=fontsize - 6,
    title="Leads",
    title_fontsize=fontsize - 4,
)

fig.tight_layout()
```

```python
import matplotlib.ticker as mticker

fontsize = 20
inferior_leads = {"ii", "iii", "avf"}  # use a set for fast membership checks

fs = 1000  # Hz
x = (start_idx + np.arange(end_idx - start_idx)) / fs  # time axis in seconds

fig, ax = plt.subplots(figsize=(15, 10))

for name, signal in zip(lead_names, channels_12):
    if name.lower() in inferior_leads:
        ax.plot(x, signal[start_idx:end_idx], label=name.upper())

ax.set_xlabel("Time (s)", fontsize=fontsize)
ax.set_ylabel("Amplitude (mV)", fontsize=fontsize)
ax.set_title("ECG Signal - Inferior Leads – Patient 001", fontsize=fontsize + 8)


# Tick styling
ax.tick_params(axis="both", labelsize=fontsize)
ax.xaxis.set_major_locator(mticker.MaxNLocator(nbins=8))
ax.xaxis.set_major_formatter(mticker.FormatStrFormatter("%.1f"))
ax.minorticks_on()

# Grid
ax.grid(True, which="major", linestyle="-", linewidth=0.8)
ax.grid(True, which="minor", linestyle=":", linewidth=0.6, alpha=0.7)

ax.legend(
    loc="upper right",
    ncol=2,
    fontsize=fontsize - 6,
    title="Leads",
    title_fontsize=fontsize - 4,
)

fig.tight_layout()
plt.show()
```
