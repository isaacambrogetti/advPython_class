**Note:** There are multiple ways to clone a repository to your local machine.

   - Do **not** place it inside another Git repository.


##### Useful Git commands

- **Typical workflow:**  
  1. Make changes, 2. `git add .` (or specific files), 3. `git commit` (with a meaningful message)

- **Synchronizing with remote:**  
  - `git push`: upload local commits to remote repository (in our case, on GitHub)
  - `git pull`: fetch and integrate remote changes  (in our case, from GitHub)

- **Working with branches:**
  Use of branches is highly recommended (e.g., one branch per feature).
  - Create a new branch and directly switch to to it: `git switch -c <branch-name>`  
  - Switch to an existing branch: `git switch <branch-name>`

- **Other useful commands:**  
  - `git stash`: temporarily store uncommitted changes (can be restored later)
  - `git log`: view commit history  
  - `git status`: view current repository state  

You may use these commands via the CLI. VS Code also provides equivalent functionality through its graphical interface.

##### Important Git considerations

- To reduce merge conflicts, avoid working on the same files simultaneously.  
- Never commit secrets (e.g., passwords, API keys).
