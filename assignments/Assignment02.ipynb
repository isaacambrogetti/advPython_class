#### Exercise 5: Number divisors *(4 points)*

A divisor of a number, is another number that divides the first number without reminder.

Example: 4 is a divisor of 20. 1 and the number itself are always divisors. 

Write a function `find_divisors` that takes a number in input and returns *all* its divisors.

Example:
```python 
find_divisors(20)  # returns [1, 2, 4, 5, 10, 20]
find_divisors(7)  # returns [1, 7]
```

````python
def find_divisors(a):
    # Empty list so numbers can be append
    divisors = []
    # Loop that adds to the list the divisors
    for i in range(1, a+1):
        if a % i == 0:
            divisors.append(i)
    return divisors

#print(find_divisors(20))
#print(find_divisors(7))
````

Implement a function `find_common_divisors` that accepts to integers and returns a sorted list of common divisors (high divisors first).

Hints: Use the function `find_divisors` from above. Sets can be helpful to implement list intersections.

````python
def find_common_divisors(a, b):
    # Find the divisors of both numbers
    div_a = find_divisors(a)
    div_b = find_divisors(b)
    
    # Convert them to sets and find the common ones
    common = set(div_a) & set(div_b)

    # Return the sorted list
    return sorted(common, reverse=True)

#find_common_divisors(14, 98)

        
````
