#### Exercise 6: DRY *(3 points)*

An important software development principle is **DRY** (Don't Repeat Yourself).  
It aims to reduce duplicated code by centralizing logic and reusing functions where possible.

1. Create a single list of tuples containing the student name and the corresponding grade.

2. Implement a function `print_result` that displays the result for a single student.

3. Create a loop to iterate through the list of tuples and use the `print_result` function to print each result.

The final output must exactly match the current output.

````python
name_1 = "Anna"
grade_1 = 5.5
print(f"Student: {name_1}, Grade: {grade_1}, Passed: {grade_1 >= 4}")

name_2 = "Ben"
grade_2 = 6
print(f"Student: {name_2}, Grade: {grade_2}, Passed: {grade_2 >= 4}")

name_3 = "Clara"
grade_3 = 3.5
print(f"Student: {name_3}, Grade: {grade_3}, Passed: {grade_3 >= 4}")
````

````python
# 1. List of tuples with the name and grade
students = [("Anna", 5.5), ("Ben", 6), ("Clara", 3.5)]

# 2. Function that displays the result for a single student
def print_result(student):
    name, grade = student
    passed = grade >= 4 # Logic evaluation to know if the student passed
    print(f"Student: {name}, Grade: {grade}, Passed: {passed}")

# 3. Loop that iterates over the list
for s in students:
    print_result(s)
````

#### Exercise 7: Sets *(2 points)*

Create code that generates **3 random integers from 1 to 6** using a loop. Add each generated integer to a set named `occurred_numbers`. Finally, print the set.

Hint: Generate the random integer with `random.randint(1, 6)`.

````python
import random

# Initialize an empty set to store unique random numbers
occurred_numbers = set()

# Loop 3 times to generate random integers
for i in range(3):
    # Generate a random integer between 1 and 6 (inclusive)
    num = random.randint(1, 6)
    # Add the number to the set
    occurred_numbers.add(num)

# Display the final set of numbers
print(occurred_numbers)
````

Has the set `occurred_numbers` always the same size?

**Your answer:** No because sets do not store duplicates. If the code generates the same number twice, the set will be smaller.

Consider a simple game: if the generated numbers are 1–2–3 (in this order), the player wins. To implement this game, what would you suggest changing in the code so far?

**Your answer:** Change the set to a list, since lists do maintain order and allow duplicates, which is needed for verifying an exact sequence.

#### Exercise 8: Loops *(2 points)*

Implement a function `average_word_length`, that accepts a list of strings and calculates the average length of all the words given as input.

Example: `print(average_word_length(["moon", "sun", "stars"]))` should output `4.0`

````python
def average_word_length(word_list):
    """
    Takes a list of words and returns the average number of characters per word.
    """
    total_length = sum(len(word) for word in word_list)
    return total_length / len(word_list)
````
